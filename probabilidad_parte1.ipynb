Axiomas Kolmogórov:
1: La probabilidad de un suceso no puede ser negativa.
2: La probabilidad de un suceso seguro es 1.
3: La probabilidad de un conjunto de sucesos excluyentes, es la suma de todas las probabilidades.

Partimos de la hipotesis de que P1 y P2 son medidas de probabilidad por lo tanto deben cumplir los axiomas:

1 Para cualquier evento A:

P(A) = a1P1(A) + a2P2(A)

Dado que P1(A) >/= 0 Y P2(A) >/= 0 y a1, a2 >/= 0, se tiene que: P(A) >/= 0.
por lo tanto P satisface la NO NEGATIVIDAD.

2 Evaluamos P(Ω):

P(Ω) = a1P1(Ω) + a2P2(Ω)

Como P1(Ω) = 1 y P2(Ω) = 1, y a1 + a2 = 1, se cumple que:

P(Ω) = a1*1 + a2*1 = a1 + a2 = 1
por lo tanto P satisface NORMALIZACIÓN

3 Sea A ∩ B = 0:

Entonces, P(A ∪ B) = a1P1(A ∪ B) + a2P2(A ∪ B)

Como P1 y P2 son medidas de probabilidad, se cumple que:

P(A ∪ B) = P1(A) + P1(B), P2(A ∪ B) = P2(A) + P2(B)

sustituyendo:

P(A ∪ B) = a1(P1(A) + P1(B)) + a2(P2(A) + P2(B))

distribuyendo:

P(A ∪ B) = a1P1(A) + a1P1(B) + a2P2(A) + a2P2(B)

reagrupando:

P(A ∪ B) = (a1P1(A) + a2P2(A)) + (a1P1(B) + a2P2(B))

lo que resulta en:

P(A ∪ B) = P(A) + P(B)
por lo tanto satisface la ADITIVIDAD

```python
import numpy as np

lambdas = np.array([1,2,3,4])
prior = np.array([0.4,0.3,0.2,0.1])
x = 4

# a) Estimación inicial del parámetro lambda_hat
lambda_hat = np.sum(lambdas*prior)
print(lambda_hat)

# b) Función de verosimilitud
Poisson = lambda lambdas,x: lambdas**x*np.exp(-lambdas)/np.math.factorial(x)
Likelihood = Poisson(lambdas,x)
print(Likelihood)

# c) Función de distribución a posteriori
Probx = np.sum(prior*Likelihood)
Posterior = (Likelihood*prior)/Probx
print(Posterior)

# d) Normalización
p_normalizada = np.sum(Posterior)
print(p_normalizada)

# e) Modelo
modelo = lambdas[np.argmax(Posterior)]
print(modelo)

# f) Mejor parámetro lambda_hat
mejor = np.sum(Posterior*lambdas)
print(mejor)
```

```python
import matplotlib.pyplot as plt
import numpy as np

def probabilidad_cumple(n):
    prob = 1.0
    for i in range(n):
        prob *= (365 - i) / 365
    return prob

n_ = np.arange(1, 81)

probabilidades = [probabilidad_cumple(n) for n in n_]

plt.figure(figsize=(10, 6))
plt.plot(n_, probabilidades, marker="o", linestyle="-", color="b", label="P(n)")
plt.title("Probabilidad de que n personas tengan cumpleaños diferentes", fontsize=14)
plt.xlabel("Número de personas (n)", fontsize=12)
plt.ylabel("Probabilidad P(n)", fontsize=12)
plt.axhline(0.5, color="red", linestyle="--", label="P(n) = 0.5")
plt.grid(alpha=0.3)
plt.legend(fontsize=12)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from itertools import permutations
from itertools import combinations
from itertools import combinations_with_replacement

States = np.array([0,1])
Prior = np.array([0.2,0.8])
T = np.array([[0.8,0.2],[0.2,0.8]])
E = np.array([[0.5,0.9],[0.5,0.1]])

Obs = np.array([1, 0, 0, 0, 1, 0, 1, 0]) # 1 = S, 0 = C
#DictH = {0: 'J', 1: 'B'}
#DictO = {0: 'C', 1: 'S'}

def GetStates(State,N):
    CStates = list(combinations_with_replacement(State,N))
    Permu = []

    for it in CStates:
      p = list(permutations(it,N))
      for i in p:
        if i not in Permu:
          Permu.append(i)

    CStates = np.array(Permu)
    return CStates

HiddenStates = GetStates(States,len(Obs))

def GetProb(T,E,Obs,State,Prior):
  n = len(Obs)
  p = 1.0

  p *= Prior[State[0]]

  for i in range(n - 1):
    p *= T[State[i + 1], State[i]]

  for i in range(n):
    p *= E[Obs[i], State[i]]

  return p

P = np.zeros(HiddenStates.shape[0])

for i in range(P.shape[0]):
  P[i] = GetProb(T,E,Obs,HiddenStates[i],Prior)

P_norm = P/np.sum(P)
ii = np.where(P == np.max(P_norm))
print(HiddenStates[ii],P_norm[ii])
plt.plot(P_norm)

ObsStates = GetStates([0,1],len(Obs))
Nobs = ObsStates.shape[0]
PObs = np.zeros(Nobs)

for j in range(Nobs):
  dim = HiddenStates.shape[0]
  P = np.zeros(dim)

  for i in range(dim):
    P[i] = GetProb(T,E,ObsStates[j],HiddenStates[i],Prior)
  PObs[j] = np.sum(P)

PObs_norm = PObs/np.sum(PObs)

ii_obs = np.where(PObs_norm == np.max(PObs_norm))
print(ObsStates[ii_obs],PObs_norm[ii_obs])
plt.plot(PObs)

np.sum(PObs_norm)
```
