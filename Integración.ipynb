```python
import numpy as np
import matplotlib.pyplot as plt

#Punto 5

def f(x):
    return np.exp(-x**2)

def trapecio(a, b, n):
    h = (b - a) / n
    x = np.linspace(a, b, n+1)
    y = f(x)
    integral = (h/2) * (y[0] + 2 * np.sum(y[1:n]) + y[n])
    return integral


a = 0
b = 1
n = int((b-a)/0.001+1)

integral = trapecio(a, b, n)
print(integral)

#Punto 6

def f(x,a):
    return np.sqrt(a**2 - x**2)/(0.5 + x)

def Trapecio(a, b, n):
    H = (b - a) / n
    X = np.linspace(a, b, n+1)
    Y = f(X,a)
    integral = (H/2) * (Y[0] + 2 * np.sum(Y[1:n]) + Y[n])
    return integral

def Simpson_3(a,b,n):
     H = (b - a) / n
     X = np.linspace(a, b, n+1)
     Y = f(X,a)
     integral = (H/3) * (Y[0] + 3* np.sum(Y[1:n]) + Y[n])
     return integral

a = -0.01
b = 0.01
n = int((b-a)/0.001+1)

integral_t = Trapecio(a, b, n)
integral_s = Simpson_3(a,b,n)
print(integral_t,integral_s)

#Punto 7
n = 100
x = np.linspace(-1, 1, n+1)
y = np.linspace(-1, 1, n+1)

X,Y = np.meshgrid(x,y)

def f(x, y):
    return np.sqrt(1 - x**2 - y**2)
z = f(X,Y)

def semiesfera(x,y,n):

  dx = (x[-1]-x[0])/n
  dy = (y[-1]-x[0])/n
  A = dx * dy

  for i in range(n+1):
    for j in range(n+1):
      a = f(x[i],y[j])
      b = f(x[i+1],y[j])
      c = f(x[i],y[j+1])
      d = f(x[i+1],y[j+1])

  integral = A*(a+b+c+d)/4

  return integral


fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, z, cmap='viridis')

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('Semiesfera de radio R = 1')

plt.show()
```
